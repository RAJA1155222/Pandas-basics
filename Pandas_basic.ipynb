Q4

What is a DataFrame in pandas and how is it different from Series?

✔ DataFrame

A DataFrame is a 2-dimensional data structure in pandas that stores data in rows and columns.

✔ Series

A Series is a 1-dimensional labeled array.

✔ Difference with example

```python
import pandas as pd

# Series (1-D)
s = pd.Series([10, 20, 30])

# DataFrame (2-D)
df = pd.DataFrame({
    "A": [10, 20, 30],
    "B": [40, 50, 60]
})
```

| Feature   | Series        | DataFrame        |
| --------- | ------------- | ---------------- |
| Dimension | 1-D           | 2-D              |
| Structure | Single column | Multiple columns |
| Example   | s             | df               |

Q7

Create a DataFrame using multiple Series.

```python
import pandas as pd

s1 = pd.Series([101, 102, 103])
s2 = pd.Series(["Amit", "Ravi", "Neha"])
s3 = pd.Series([20, 21, 22])

df = pd.DataFrame({
    "Roll No": s1,
    "Name": s2,
    "Age": s3
})

print(df)
```
